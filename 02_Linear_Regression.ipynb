### import the important libraries


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### read the data

```python
df = pd.read_csv(r'boston.csv')
df.head()
```

### get information of the data

```python
df.info()
```

### we will split the data

```python
x = df.drop('MEDV',axis = 1) # independent features
y = df['MEDV'] # dependent
```

### train test and split


```python
#train test Split


from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

# Apply Linear Regression

```python
from sklearn.linear_model import LinearRegression

# initiaze  

lr = LinearRegression()

# train the model

linear_model = lr.fit(x_train,y_train)

# prediction of y

y_pred = linear_model.predict(x_test)
```

```python
y_pred
```

```python
# Model evaluation
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
mse = mean_squared_error(y_test,y_pred)
print(f'Mean Square error of LR is : {mse}')
```

```python
mae = mean_absolute_error(y_test,y_pred)
print(f'Mean Absolute error of LR is : {mae}')
```

```python
rmse = np.sqrt(mse)
print(f'Root Mean Square Error of LR is : {rmse}')
```

```python
r_square = r2_score(y_test,y_pred)
print(f'R_Square of LR is : {r_square}')
```

# Plotting the Actual Values and Predicted values

```python
plt.figure(figsize = (10,6))
plt.scatter(y_test,y_pred)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title("Actual Vs Predicted Prices")

plt.show()
```

# It will show the nearest data point from the Zero

```python
linear_model.coef_
```

# getting coefficients

```python
coefficients = pd.DataFrame({
    'Features': x.columns,
    "Coefficients": linear_model.coef_  
}).sort_values(by = 'Coefficients',ascending = False)
print(coefficients)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x= 'Coefficients',y='Features',data = coefficients)
plt.title("Feature Important in Predicting Housing Prices")

plt.show()
```
